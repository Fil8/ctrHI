```python
import numpy as np
from astropy.coordinates import Angle
from astropy.io import fits
from astropy import units as u 
from ants import absInt, fluxInt, headPlay, fitsPlay #radiobs modules


fl = fluxInt.fluxint()
hp = headPlay.headplay()
fp = fitsPlay.fitsplay()


get_ipython().magic(u'pylab inline')


home = '/Users/'
#home = '/home/'


rootDir = home+'maccagni/Projects/MFS/FornaxA/centreHI/'
cubeDir = home+'maccagni/data/MFS/MeerKATComm/fa4/cubes/'
acaDir = home+'maccagni/data/MFS/FornaxA/ACA/'

print('''\n\t+---------+\n\t Inputs loaded\n\t+---------+''')
```

```python
#------------------#
# CO BUNIT OUTDATED -> new convolved moment map already in Jy/beam*km/s
#------------------#

coMomName = acaDir+'/co10Cubes/NGC1316_TP_7M_CO.feather.image_mom0.fits.gz'
momDirOut = rootDir+'moments/'

momOutName=momDirOut+'mom0_CO_kana.fits'

coCubeName = rootDir+'cubes/CO_vrad.fits'
coHead=hp.printHead(coCubeName)

fileIn = fits.open(coMomName)
datas = fileIn[0].data
heads  = fileIn[0].header
datas = np.multiply(datas,float(2.6005929685193367e-06))
heads['BUNIT'] = 'Jy/beam*km/s'
del heads['DATAMIN']
del heads['DATAMAX']
fits.writeto(momOutName,datas,heads,overwrite=True)

print '''\n\t+---------+\n\t BUNIT modified\n\t+---------+'''
```

```python
#------------------#
# CO surface brightness
#------------------#

momOutKms=momDirOut+'mom0_CO_kana.fits'


D_L = 20.8
nu_obs = 114.619317
z=0.00485
C = 2.99792458e10
kb = 1.380649e-16
Jy = 1e-23
asecSr = 4.25e10

fac = 2.*np.log(2.)/(np.pi)*(C*C)/(kb)*asecSr*Jy/1e18

print('CO brightnes temperature factor = \t{0:e}'.format(fac))                                             


coHead=hp.printHead(momOutKms)

bmaj = 15.175263
bmin = 7.734163

beamAreaFreq = bmaj*bmin*nu_obs*nu_obs
print('BeamArea Freq = \t{0:e}'.format(beamAreaFreq))                                             

fac = fac/beamAreaFreq

facMassPc =4.3

#z = 0.00587#NED:
#inputMom = momOutName # Jy/beam*m/s

kMom=momDirOut+'mom0_CO_kelvin_kana.fits'
surfBright=momDirOut+'surfBright_CO_kana.fits'


fp.multVal(momOutKms,fac,kMom)
hp.putHead(kMom,'BUNIT','K*km/s')


fp.multVal(kMom,facMassPc,surfBright)
hp.putHead(surfBright,'BUNIT','Msun pc-2')                                                
                                                
print '''\n\t+---------+\n\t Moments Converted to Surface Brightness\n\t+---------+'''
                                             
```
